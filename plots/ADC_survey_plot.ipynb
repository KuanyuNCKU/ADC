#Plotting ADC Survey Data

```python
import matplotlib.pyplot as plt
import pandas as pd

data = "https://github.com/bmurmann/ADC-survey/blob/main/xls/ADCsurvey_rev20220719.xls?raw=true"
df1 = pd.read_excel(data, sheet_name='ISSCC')
df2 = pd.read_excel(data, sheet_name='VLSI')
df = pd.concat([df1, df2])
df
```

```python
print(df.keys())
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
ax.scatter(x = df['SNDR_plot [dB]'], y = df['P/fsnyq [pJ]'], marker='s', s=100)
ax.set_yscale('log')
ax.set_ylim((1e-1, 1e7))
plt.xlabel("SNDR [dB]")
plt.ylabel("P/fsnyq [pJ]")
plt.grid(True)
plt.show()
```
